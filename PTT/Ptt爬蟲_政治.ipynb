```python
title, time, content, comment = [], [], [], []
for n in tqdm_notebook(range(4702,4699,-1)):
    res = requests.get("https://www.ptt.cc/bbs/HatePolitics/index"+str(n)+".html")
    soup = BeautifulSoup(res.text)
    for i in soup.select(".title"):
        try:    
            if i('a') == []:
                continue
            else:
                res2 = requests.get("https://www.ptt.cc"+i('a')[0]['href'])
                soup2 = BeautifulSoup(res2.text)
                title.append(soup2.select(".article-meta-value")[2].text)
                time.append(soup2.select(".article-meta-value")[3].text)
                content.append(soup2.select("#main-content")[0].text.split("※")[0].split(time[-1])[1])
                lst = []
                for m in soup2.select(".push .f3.push-content")[0:]:
                    lst.append(m.text)
                comment.append(lst)
        except IndexError:
            print("出錯喔")
```

```python
ptt_HatePolitics = list(zip(title, time, content,comment))
```

```python
import pandas as pd
df = pd.DataFrame(ptt_HatePolitics, columns=['標題','時間','內文','留言'])
```
